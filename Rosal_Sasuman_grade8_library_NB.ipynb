```python
url = "https://raw.githubusercontent.com/KobeKarlMSasuman/Rosal-Sasuman-grade8-library-Repo/refs/heads/main/books.json"

response = requests.get(url)
books_data = response.json()

print("JSON Successfully Loaded:")
print(books_data)
```

```python
cred = credentials.Certificate("firebase_key.json")

firebase_admin.initialize_app(cred, {
    'databaseURL': 'https://rosal-sasuman-library-nb-default-rtdb.asia-southeast1.firebasedatabase.app/'
})
```

```python
ref = db.reference("books")

ref.set(books_data["books"])

print("Books uploaded successfully!")
```

```python
books = ref.get()

print("All Books Data:")
print(books)
```

```python
print("\nBook Titles:")
for book in books:
    print(book["title"])
```

```python
print("\nTitle and Availability:")
for book in books:
    print(book["title"], "-", book["details"]["available"], "available")
```

```python
total_books = 0

for book in books:
    total_books += book["details"]["copies"]

print("\nTotal number of books in library:", total_books)
```
